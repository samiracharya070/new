### JOINING DATA WITH PANDAS

**Create two data that join** 

```python
import pandas as pd

data1 = {
    'name': ['Ram', 'Samir', 'Pawan', 'Sita', 'Rita'],
    'rollno': [1, 2, 3, 4, 5],
    'age': [20, 21, 19, 22, 24],
    'height': [160, 170, 165, 180, 158]
}
df1 = pd.DataFrame(data1)


data2 = {
    'rollno': [1, 2, 3, 4, 5],
    'total_marks': [450, 380, 290, 410, 320],  # out of 500
    'percentage': [90.0, 76.0, 58.0, 82.0, 64.0],
    'grade': ['A', 'B', 'C', 'A', 'B']
}
df2 = pd.DataFrame(data2)

# Output the DataFrames
print(f"DataFrame 1:{df1.shape} \n{df1}")
print(f"\nDataFrame 2:{df2.shape} \n{df2}")

```

**Inner join ( common join )**   
   If we want to merge the DataFrames, we need at least one common column between them.  
   here rolllno column are common

```python
#REMOVE COLUMN
# df1 = df1.drop(columns=['rollno'])
# df2 = df2.drop(columns=['rollno'])

# print(f"DataFrame 1: {df1.shape} \n{df1}")
# print(f"\nDataFrame 2: {df2.shape} \n{df2}")


inner_join = df1.merge(df2 )
print(f"\nDataFrame 2: {inner_join.shape} \n{inner_join}")

#  add suffixes into different table 
#  if both table have same column name more then one only suffixes are used 
df2['age'] = [20, 22, 19, 21, 24]
inner_join_suff = df1.merge(df2 , on ='rollno', suffixes =('_1','_2'))
print(f"\nDataFrame 2: {inner_join_suff.shape} \n{inner_join_suff}")
```

**One to many relationships**  
the number of row into two table are different 

```python
new_data = pd.DataFrame([{'rollno': 10,'age': 23,'total_marks': 400,'percentage': 80.0,'grade': 'B'}])
df2 = pd.concat([df2 , new_data],ignore_index=True)
print(df2)

one_to_many = df1.merge(df2 ,on ='rollno')
print(f"\nDataFrame 2: {one_to_many.shape} \n{one_to_many}")
```

**Left join**  
the table of left is same order and the right table is arrange according to th age in left table

```python
left_join = df1.merge(df2 ,on = 'age',how ='left')
print(left_join)
# df2['age'] = [20, 22, 19, 21, 24]
#   'age': [20, 21, 19, 22, 24],
# here age 21 in left table match with rollno 4 in right table 
```

**Right merge**  
The table of right is same order and the left table is arrange according to th age in right table 

```python
right_join = df1.merge(df2 ,on ='rollno' ,how= 'right')
print(right_join)
#   df2['age'] = [20, 22, 19, 21, 24]
#   'age': [20, 21, 19, 22, 24],

right_join_on = df1.merge(df2 ,left_on ='age' ,right_on= 'rollno' ,how= 'right')
print(f"\n{right_join_on}")

# bec
```

**Outer join** 

```python
outer_join = df1.merge(df2 , on ='rollno', how = 'outer')
print(outer_join)
```

**Self join  on rolllno**  
  self join on rollno and age is empty 

```python
self_join = df1.merge(df1 ,left_on='rollno',right_on='rollno')
print(self_join)

self_join_e = df1.merge(df1 ,left_on='rollno',right_on='age')
print(self_join_e)
```

**Merging on indexes**    
   Merging on the basic of the index of the table 

```python
m_index1 = df1.set_index('rollno')
m_index2 = df2.set_index('rollno')

# merge on basic of rollno
merge_index = m_index1.merge(m_index2, right_index=True,left_index=True)
print(merge_index)
```

**Filtering joins**   
  **semi join** : only data that have in both table are present

```python
semi_join = df1.merge(df2,on='rollno')
print(semi_join)
print(df1['rollno'].isin(semi_join['rollno']))
print(df2['rollno'].isin(semi_join['rollno']))
print(df1[df1['rollno'].isin(semi_join['rollno'])])
```

**Anti merge** :  _print the data only present in one table in may be left and right_

```python
anti_join = df1.merge(df2,on ='rollno',how='right',indicator=True)
print(f"\n{anti_join}")
anti_join_right = anti_join.loc[anti_join['_merge']=='right_only','rollno']
print(f"\n{anti_join_right}")

# data only present in the right table is print
print(f"\n{df2[df2['rollno'].isin(anti_join_right)]}")
```

**Concatenate DataFrames together verticallys**  

```python

data3 = {
    'name': ['Amit', 'Neha', 'Vikas', 'Meena', 'Rohit'],
    'rollno': [6, 7, 8, 9, 10],
    'age': [23, 20, 22, 21, 25],
    'height': [172, 160, 178, 165, 169]
}
df3 = pd.DataFrame(data3)

# print(df3)


ver = pd.concat([df1,df3],keys=['df1','df3'])
print(f"\n{ver}")   # index of the table are wrong aline

ver_join = pd.concat([df1,df3],ignore_index=True)
print(f"\n{ver_join}")
```

**Using merge_ordered()**

```python
pd.merge_ordered(df1,df2,on='rollno',how='right')
```

```python
#  fill_methos ffill fill the nan with uper value or forward value
pd.merge_ordered(df1, df2, on='rollno', how='right', fill_method='ffill')
```

**Using merge_asof()**  
Similar to a merge_ordered() left join Similar features as merge_ordered()  
left keys and righy keys must be sorted

```python
pd.merge_asof(df1, df2, on='rollno')
```

**Selecting data with.query()**

```python
df2.query('percentage >80')
```

```python
pd.merge_asof(df1, df2, on='rollno').query('percentage >80')
```

```python
pd.merge_asof(df1, df2, on='rollno').query('grade =="A" or grade=="B"')
```

**Reshaping data with.melt()**
