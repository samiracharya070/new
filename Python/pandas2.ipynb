Add a new column in data frame 

```python
pd_data["total"] = pd_data["ENG"]+pd_data["MATH"]+pd_data["SCI"]+pd_data["NEP"]
# pd_data["total"] = pd_data[["ENG", "MATH", "SCI", "NEP"]].sum(axis=1)

pd_data["percentage"] = (pd_data["total"] / 400) *100
print(f"{pd_data}\n")
print(pd_data.sort_values("total",ascending=False))

#  save the data to a new csv file
pd_data.to_csv('data3.csv')
```

DATA MANIPULATION WITH PANDAS   
 Grouped summarystatistics   
   data_name.groupby("name")["MATH"].mean()


```python
# to get the data for special element
print(pd_data[pd_data['name']=="Bikash Lama"]['MATH'].mean())

# get the data for the special columns
print(pd_data.groupby("name")["MATH"].mean())

print(f"\n{pd_data.groupby("name")["MATH"].agg(['mean','sum','max'])}")

print(f"\n{pd_data.groupby(["name","rollno"])[["ENG","MATH","SCI"]].mean()}")
```

Explicit indexes

```python
#  set name as index in data frame 
pd_data_ind = pd_data.set_index('name')
print(pd_data_ind)
print(f"\n{pd_data_ind.loc['Riya Shrestha']}")
print(f"\n{pd_data_ind.reset_index()}")
print(f"\n{pd_data_ind.reset_index(drop=True)}")
```

Indexes make subsetting simples

```python
print(pd_data[pd_data['name'].isin(['Sita Karki','Bikash Lama'])])
print(f"\n{pd_data_ind.loc[['Sita Karki','Bikash Lama']]}")
```

```python
# set rollno as index 
rollno_ind = pd_data.set_index('rollno')
print(f"\n{rollno_ind.loc[[1,3,5,7]]}")

# sort the data according to index 
print(rollno_ind.sort_index(ascending=False))
```

Multi-level indexes a.k.a. hierarchical indexes   
    give first index ao give (first ,second)index

```python
name_roll = pd_data.set_index(['name','rollno'])
print(name_roll.loc[['Niraj Thapa','Aarav Sharma']])
print(name_roll.loc[[('Niraj Thapa',3),('Kiran Gurung',7)]])
```
