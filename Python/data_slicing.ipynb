### Data Slicing or Subset

**Data slicing or subset using columns**

```python
import pandas as pd

df = pd.read_csv('data_stu.csv')
print("Data slicing or subset using columns")

print(df['Student_Name'])
print(df['Roll_No'])
```

```python
print(df[0 :1])
new = df.set_index('Student_Name')


print(new[0:4])
print(new.loc['Ram Thapa':'Dipesh Rai'])
print(new.iloc[0:5])
```
