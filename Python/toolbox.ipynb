**Iterator**   
Produces next value with next()


**iterating over iterables:next()**

```python
word = 'Data'
it = iter(word)
print(next(it))
print(next(it))
print(next(it))\
    
ite = iter(word)
print(*ite)
```

**Using enumerate()**

```python
fruits = ['apple', 'banana', 'cherry', 'mango', 'orange']
e=  enumerate(fruits)
print(type(e))
print(e)
e_list = list(e)
print(e_list)

for key ,value in enumerate(fruits):
    print(key , value)
```

**Using zip()**

```python
start_f =['a', 'b', 'c', 'm', 'o']
z= zip(start_f,fruits)
print(type(z))
print(*z)

z= zip(start_f,fruits)
z_list = list(z)
print(z_list)
# z_dict = dict(zip(start_f, fruits))
# print(z_dict)


for z1,z2 in zip(start_f,fruits):
    print(z1,z2)
```

**List comprehensions**

```python
num=[1,3,5,7,9]
new_num= [num+1 for num in num]
print(num)
print(new_num)

result = [num for num in range(10)]
print(result)
```

```python
new_1 = []
for num1 in range(0,2):
    for num2 in range(4,6):
        new_1.append((num1 , num2))
        
print(new_1)
```

```python
new_2 = [(num1,num2) for num1 in range(0,2) for num2 in range(4,6)]
print(new_2)
```

**Advanced comprehension()**

```python
new_3 =[num*2 for num in range(10) if num % 2 == 0]
print(new_3)
new_4 =[num*2 if num % 2 == 0 else num for num in range(10) ]
print(new_4)
```
