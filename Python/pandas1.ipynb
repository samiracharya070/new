Locate Row

Pandas use the loc and iloc attribute to return one or more specified row(s)

```python
print(pd_data.index)         # ptint row of the data 

print(pd_data.iloc[0])       # access the data of index 0 
print("  ")
print(pd_data.loc['AF'])     # access thr data of row  AF
print("  ")
print(pd_data.iloc[1,2])     # access the data in [row ,column]

# in data formate 
print(pd_data.iloc[[4,3,2]])  # access the more then one data 
print(pd_data.loc[['NE']])    # access the data using the index row NE
```

subset the data 

```python
new_data1 = pd_data.iloc[: ,[0,3]]          # Select all rows and only columns at index 0 and 3 (specific columns)
new_data2 = pd_data.iloc[: ,0:1]            # Select all rows and column index 0 only (note: 0:1 is a slice, returns a DataFrame, not a Series)
new_data3 = pd_data.iloc[0:5 ,[0,1,2]]      # Select rows from index 0 to 4 (5 is excluded), and columns at index 0, 1, and 2
new_data4 = pd_data.iloc[0:3,0:5]           # Select rows from index 0 to 2 (3 is excluded), and columns from index 0 to 4 (5 is excluded)

print(f"pd_data:\n{pd_data}\n")         
# print(f"new_data1:\n{new_data1}\n")
# print(f"new_data2:\n{new_data2}\n")
print(f"new_data3:\n{new_data3}\n")
print(f"new_data4:\n{new_data4}\n")
```
