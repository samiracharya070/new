Creating Data Frames   
 
 1 . From a list of dictionaries     
       Constructed row by row 

 2 . From a dictionary of lists   
      Constructed column by column


```python
import pandas as pd
data1 = [{'name':'samir','age':20,'height':5.6},
        {'name':'ram','age':19,'height':5.5}
        ]
pd_data1 = pd.DataFrame(data1)
print(pd_data1)
```

```python
data2 ={'name':['samir','ram','shyam'],
        'age':[20,19,22],
        'height':[5.6,5.5,8]}
pd_data2 = pd.DataFrame(data2)
print(pd_data2)
```

Removing missing values in data frame

```python
import numpy as np


data = {
    'name': ['samir', 'ram', 'hari', 'sita', 'bikash', 'anita','pawan'],
    'age': [20, 19, np.nan, 21, 23, 20, 24],
    'height': [5.6, np.nan, 5.2, 5.4, 5.7, np.nan, 6.1],
    'weight': [60, 58, 72, np.nan, 65, np.nan, 75]  # new column
}

df = pd.DataFrame(data)
print(df)
print(f"\n{df.isna()}")
print(f"\n{df.isna().any()}")
print(f"\n{df.isna().sum()}")


```

Removing missing values

Replacing missing values

```python
#  Removing missing values
print(df.dropna())
#   Replacing missing values by 0
print(f"\n{df.fillna(0)}")
```

Plot a graph of missing values 

```python
import matplotlib.pyplot as plt

print(df.isna().sum())
df.isna().sum().plot(kind='bar')
plt.show()
```
