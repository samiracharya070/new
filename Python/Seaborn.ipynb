```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data_asia.csv')
print(df.columns)
Population= df['Population']

sns.scatterplot(x='Country', y='GDP', hue='Rank', data=df)
plt.xticks(rotation=90)
plt.show()
```

**seprte the graph into two row**

```python
df_2 = pd.read_csv("data_p.csv")

sns.relplot(x='name',y='age',data=df_2,kind="scatter",hue='sex',row="sex")
plt.xticks(rotation = 90)
plt.show()
```

**seprate the graph into column**

```python

# Set the theme to 'whitegrid'
sns.set_theme(style="whitegrid")
g = sns.relplot(
    x='name', 
    y='age', 
    data=df_2, 
    kind="scatter", 
    col='sex',  # fixed 'sex '
    col_wrap=2,
    size='age'   # optional, ensure 'age' is numeric
)

# Rotate x-axis labels in all facets
g.set(xlabel=' name ', ylabel='age')  
g.set_xticklabels(rotation=90)

plt.tight_layout()
plt.show()
```

size = size of line ,scatter point according to the column  
sizes = acording to no of variabe in size (first ,second)  
style = style of line ,scatter point according to the column   

```python
sns.set_theme(style="white")
g= sns.relplot(x='name',y='age',data=df_2,kind='line',size='sex',hue='sex',sizes=(2,1),style= 'sex')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_theme(style="whitegrid")
g= sns.relplot(x='name',y='age',data=df_2,kind='scatter',style= 'age')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_theme(style="whitegrid")
g= sns.relplot(x='name',y='age',data=df_2,kind='line',style= 'age',markers=True)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_palette="RuBu"
g= sns.countplot(x='age',data=df_2,hue='sex')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_style="white"
g = sns.catplot(
    x='name',
    y='age',
    data=df_2,
    kind='swarm',
    hue='sex'  # color by 'sex'
)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_style="ticks"
g = sns.catplot(
    x='name',
    y='age',
    data=df_2,
    kind='point',
    hue='sex'  # color by 'sex'
)
plt.xticks(rotation=90)
plt.show()
```
