To import Matplotlib (specifically its plotting module pyplot)

```python
import matplotlib.pyplot as plt
```

Here two list one in years and another is population on that year
The population values in the list I provided are in billions.

```python
years = [1995, 1998, 2001, 2004, 2007, 2010, 2013, 2016]
population = [5.7, 5.9, 6.1, 6.4, 6.7, 6.9, 7.1, 7.4]

plt.plot(years,population )
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()

```

```python
plt.scatter(years ,population )
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()

```

```python
import numpy as np


# Generate synthetic data from two normal distributions
data1 = np.random.normal(loc=50, scale=5, size=500)
data2 = np.random.normal(loc=70, scale=10, size=500)
data = np.concatenate((data1, data2))

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Combined Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
