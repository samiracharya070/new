### Matplotlib 

Read the data from the csv file 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data2.csv')
print(df)
```

```python
fig , ax =plt.subplots()
ax.plot(df['rollno'],df['ENG'],label='eng')
ax.plot(df['rollno'],df['MATH'],label='math')
plt.legend()
plt.show()
```

#### Customizing your plots

```python
fig , ax =plt.subplots()
ax.plot(df['rollno'],df['NEP'],label='Nep', marker ='*',linestyle=':')
ax.plot(df['rollno'],df['SCI'],label='Sci',marker='x',linestyle='--')
ax.set_xlabel("rollno")
ax.set_ylabel('marks in nep and sci')
plt.grid()
plt.legend()
plt.show()
```

***Small multiples***

```python
fig , ax =plt.subplots(2,2 , figsize =(12,8))
ax[0,0].plot(df['rollno'],df['ENG'],label='eng')
ax[0,1].plot(df['rollno'],df['MATH'],label='math')
ax[1,0].plot(df['rollno'],df['NEP'],label='Nep')
ax[1,1].plot(df['rollno'],df['SCI'],label='Sci',marker='*',linestyle='none')
plt.grid()

for axes in ax.flat:
    axes.set_xlabel("rollno")
    axes.set_ylabel("marks")

plt.show()
```

```python
fig , ax =plt.subplots()
ax.bar(df['rollno'],df['ENG'])
ax.bar(df['rollno'],df['MATH'],bottom=df['ENG'])

plt.show()
```

**import new csv data file that have student name and total marks in 3 exam**

```python
df2 = pd.read_csv('data_stu.csv',index_col=0)

fig, ax = plt.subplots()
ax.plot(df2['Roll_No'],df2['Total_Marks_1'])
ax.set_xlabel("Roll No")
ax.set_ylabel('Marks in first exam')
ax.set_xticks(df2['Roll_No'])
plt.show()
```

```python


fig, ax = plt.subplots()
ax.plot(df2['Roll_No'],df2['Total_Marks_1'])
ax.set_xlabel("Roll No")
ax.set_ylabel('Marks in first exam',color='blue')
ax.set_xticks(df2['Roll_No'])


ax2= ax.twinx()
ax2.plot(df2['Roll_No'],df2['Total_Marks_2'],color='red',linestyle='--')
ax2.set_ylabel('Marks in second exam',color='red')
ax2.tick_params('y', colors='red')


plt.show()
```

```python

max_mark = df2['Total_Marks_2'].max()
roll_with_max = df2[df2['Total_Marks_2']==max_mark]['Roll_No'].values[0]


fig, ax = plt.subplots()

# Plot both lines on the same axis
ax.plot(df2['Roll_No'], df2['Total_Marks_1'], label='Exam 1', color='blue')
ax.plot(df2['Roll_No'], df2['Total_Marks_2'], label='Exam 2', color='red', linestyle='--')

ax.annotate('Highest Marks',
            xy=(roll_with_max, max_mark),
            xytext=(roll_with_max + 1, max_mark + 2),
            arrowprops=dict(facecolor='black', arrowstyle='->'),
            fontsize=9)


ax.set_xlabel("Roll No")
ax.set_ylabel("Marks (both exams)")
ax.set_xticks(df2['Roll_No'])
ax.legend()  # Show legend to distinguish lines
plt.grid()
plt.show()
```

```python
# Let's say we know someone scored 97 in Exam 1
target_marks = df2['Total_Marks_2'].max()

# Get the row(s)
result = df2[df2['Total_Marks_2'] == target_marks]['Roll_No'].values[0]
print(result)
```

**Plot graph using Bar**

```python
fig , ax=plt.subplots()

ax.bar(df2.index,df2['Total_Marks_1'])
ax.set_xlabel("Student Name")
ax.set_ylabel("Marks in Exam 1")
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(df2.index, df2['Total_Marks_1']) 
ax.set_xlabel("Student Name")
ax.set_ylabel("Marks in Exam 1")
ax.set_title("Bar Chart of Marks in Exam 1")
plt.xticks(rotation=90)  # Rotate for better readability
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(df2.index, df2['Total_Marks_1'], label='Exam 1')  
ax.bar(df2.index, df2['Total_Marks_2'],bottom=df2['Total_Marks_1'], label='Exam 2') 
ax.bar(df2.index, df2['Total_Marks_3'],bottom= df2['Total_Marks_2'] + df2['Total_Marks_1'], label='Exam 3') 
ax.set_xlabel("Student Name")
ax.set_ylabel("Marks in Exam 1")
ax.set_title("Bar Chart of Marks in Exam 1")
plt.xticks(rotation=90)  # Rotate for better readability
plt.tight_layout()
# plt.legend()
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist( df2['Total_Marks_1'],bins=20,histtype='step') 
ax.hist( df2['Total_Marks_2'],bins=20,histtype='step') 
ax.set_xlabel("Student Name")
ax.set_ylabel("Marks in Exam 1 and 2")
ax.set_title("Bar Chart of Marks in Exam 1 and 2")
plt.xticks(rotation=90) 
plt.tight_layout()
plt.show()

```

```python
fig,ax =plt.subplots()

ax.bar('marks in exam 1',df2['Total_Marks_1'].mean(),yerr=df2['Total_Marks_1'].std())
ax.bar('marks in exam 2',df2['Total_Marks_2'].mean(),yerr=df2['Total_Marks_2'].std())
ax.bar('marks in exam 3',df2['Total_Marks_3'].mean(),yerr=df2['Total_Marks_3'].std())
plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(df2['Roll_No'], df2['Total_Marks_1'],label='exam 1',color='red') 
ax.scatter(df2['Roll_No'], df2['Total_Marks_2'],label='exam 2',color='green') 
ax.scatter(df2['Roll_No'], df2['Total_Marks_3'],label='exam 3',color='blue') 
ax.legend()
ax.set_xlabel("Roll no")
ax.set_ylabel("Marks ")
ax.set_xticks(df2['Roll_No'])
plt.tight_layout()
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(df2['Roll_No'], df2['Total_Marks_1'],c=df2['Total_Marks_1']) 

ax.set_xlabel("Roll no")
ax.set_ylabel("Marks ")
ax.set_xticks(df2['Roll_No'])
plt.tight_layout()
plt.show()
```
