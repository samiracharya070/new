```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV
df = pd.read_csv('data_asia.csv')

# Extract data
countries = df['Country']
populations = df['Population']

# Plotting: marker size scaled by population (e.g., divide to control visual scale)
plt.figure(figsize=(10, 6))
plt.scatter(countries, [1]*len(countries), s=populations / 1000000, alpha=1, color='skyblue', edgecolors='black')

# Enhancements
plt.title(' Countries Population (2020)')
plt.xlabel('Country')
plt.yticks([])  # Hide y-axis values

plt.tight_layout()

# Add population labels

plt.xticks(rotation=90)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('data_asia.csv')

# Extract columns
countries = df['Country']
populations = df['Population']
gdps = df['GDP']

# Create scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(countries, gdps, s=populations / 1000000, color='skyblue', alpha=0.7, edgecolors='black')

# Labeling
plt.title('SAARC Countries: GDP vs Population (2020)', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('GDP (USD)', fontsize=12)
plt.xticks(rotation=45)

# Annotate GDP values
for i in range(len(df)):
    plt.text(countries[i], gdps[i] + 2e10, f"${gdps[i]/1e9:.2f}B", ha='center', fontsize=8)

plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
