### Statistics in Python

***import pandas , matplotlib,numpy***  
Read the data from the csv file

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data_stu.csv', index_col=0)
print(df)
df['Total_Marks'].hist(bins=20 ,rwidth= 0.8)
plt.show()
```

**Calculate mean ,median,mode**

```python
import statistics as st
print(np.mean(df['Total_Marks']))
print(np.median(df['Total_Marks']))
print(st.mode(df['Total_Marks']))

df['Total_Marks'].agg(['mean', 'median'])
```

```python
mean_value = np.mean(df['Total_Marks'])
median_value = np.median(df['Total_Marks'])
plt.hist(df['Total_Marks'],bins=15,color = 'lightgreen',edgecolor='black')

plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='orange', linestyle='dotted', linewidth=2, label=f'Median: {median_value:.2f}')



plt.title('Histogram with Mean and Median')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

```

**variance and standar daviation**

```python
print(f"{np.var(df['Total_Marks'],ddof=1)} \n{np.std(df['Total_Marks'],ddof=1)}")
```

**Quantile**  
The function `np.quantile()` is used to calculate a specific quantile (or percentile) of a dataset. In the line:  
Quantile	Meaning  
0.0	   ->   Minimum value  
0.25	->  First quartile (Q1)  
0.5	    ->  Median (Q2)  
0.75   	->  Third quartile (Q3)  
1.0    	->  Maximum value   

```python
print(f" {np.quantile(df['Total_Marks'],0.5)} \n {np.median(df['Total_Marks'])}")
print(f" {np.quantile(df['Total_Marks'],[0,0.25,0.5,0.75,1])} ")
print(f" {np.quantile(df['Total_Marks'],np.linspace(0,1,9))} ")   # linspace(0,1 totla_no div -1 )

np.quantile(df['Total_Marks'], 1) - np.quantile(df['Total_Marks'], 0)
```

```python
print(f"\n{df.count()}")
print(f"\n{df.sum()}")

print(f"\n{df['Total_Marks'].count()}")
print(f"\n{df['Total_Marks'].sum()}")
```

```python
print(f"\n{df.sample(9)}")                  # data can not repeat
print(f"\n{df.sample(9 ,replace=True)}")    # data can repeat
```

###  Probability

```python
import pandas as pd

# Create the dice DataFrame
dice = pd.DataFrame({
    'number': [1, 2, 3, 4, 5, 6],
    'prob': [1/6] * 6
})

print(dice)
# throw the dice 10 times and get the data  
dice_10 = dice.sample(10,replace=True)
# print(dice_10)
print(dice_10['number'].mean())
print(dice['number'].sample(10000,replace=True).mean())
plt.hist(dice_10['number'], bins=[1,2,3,4,5,6,7], align='left', rwidth=0.8)
plt.show()

prob_10 = dice.sample(10,replace=True)
print((prob_10['number']==4).sum())

```
