DATA MANIPULATION WITH PANDAS   
 Grouped summarystatistics   
   data_name.groupby("name")["Sales"].mean()

```python

import pandas as pd

# Sample data frame
data = {
    'Store': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Product': ['Apples', 'Bananas', 'Oranges',
                'Apples', 'Bananas', 'Oranges',
                'Apples', 'Bananas', 'Oranges'],
    'Month': ['Jan', 'Jan', 'Jan',
              'Feb', 'Feb', 'Feb',
              'Mar', 'Mar', 'Mar'],
    'Sales': [100, 150, 80,
              120, 160, 90,
              130, 170, 100]
}

df = pd.DataFrame(data)
print(df.drop_duplicates(subset='Store'))
# print(df)
print(f"\n{df[df['Store']=='A']['Sales'].sum()}")
print(df.groupby('Store')["Sales"].sum())
data1 = df.groupby('Store')["Sales"]
print(data1.mean())
print(f"\n{data1.agg(['sum','mean'])}")
print(f"\n{df.groupby(['Store','Month'])['Sales'].sum()}")

```

Pivot tables  
  Group by to pivot table


```python
# import numpy as np
print(df.pivot_table(values="Sales",index='Store'))

# print(df.pivot_table(values='Sales',index='Store',aggfunc=np.mean))
print(df.pivot_table(values='Sales',index='Store',aggfunc='mean'))
print(f"\n{df.pivot_table(values='Sales',index='Store',aggfunc=['mean','sum'])}")

print(f"\n{df.pivot_table(values='Sales',index='Store',columns='Month',aggfunc='mean')}")
print(f"\n{df.pivot_table(values='Sales',index='Store',columns='Month',aggfunc='mean' ,fill_value = 0)}")
print(f"\n{df.pivot_table(values='Sales',index='Store',columns='Month',aggfunc='mean' ,fill_value = 0 ,margins= True)}")
```
