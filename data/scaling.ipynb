**Standardizing Scaling**   
Standardizing scaling is a technique used in data preprocessing to transform features so that they have a mean of zero and a standard deviation of one.    
standardizing scaling formula:   
x' = (x - μ) / σ   
where:   
x = original value   
μ = mean of the feature   
σ = standard deviation of the feature   
x' = standardized value (mean 0, standard deviation 1)   

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[['total_score']])
```

```python
ss.transform(dataset[['total_score']])
```

```python
dataset['total_score_ss'] = pd.DataFrame(ss.transform(dataset[['total_score']]),columns=['new'])
dataset.head(6)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('before')
sns.histplot(dataset['total_score'])
plt.subplot(1,2,2)
plt.title('after')
sns.histplot(dataset['total_score_ss'])
plt.show()
```

**Min-Max Scaling**
Min-Max Scaling, also known as normalization, is a technique used to scale data to a fixed range, typically [0, 1].    
The formula for Min-Max Scaling is:   
x' = (x - x_min) / (x_max - x_min)   
where:   
x = original value   
x_min = minimum value in the feature   
x_max = maximum value in the feature   
x' = scaled value in the range [0, 1]   


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms= MinMaxScaler()
arr = mms.fit_transform(dataset[['attendance_percentage']])
arr
```

```python
dataset['attendance_percentage_mms'] = pd.DataFrame(arr)
dataset.head(6)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('before')
sns.histplot(dataset['attendance_percentage'])
plt.subplot(1,2,2)
plt.title('after')
sns.histplot(dataset['attendance_percentage_mms'])
plt.show()
```
