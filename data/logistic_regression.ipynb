```python
dataset['en_data_1'] = pd.get_dummies(dataset['pass'], drop_first=True, dtype='int64')
dataset.head(6)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
new_data = ohe.fit_transform(en_data.to_frame()).toarray()
```

```python
en_data_2 = en_data_2.drop(columns='yes')
en_data_2.head(6)
```

```python
dataset['en_data_2'] = en_data_2
dataset.head(6)
```

```python
sns.scatterplot(x='total', y='en_data_2', data=dataset)
plt.show()
```

```python
x= dataset[['total']]
y= dataset['en_data_2']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr =LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test) 
```

```python
plt.figure(figsize=(7,4))
sns.scatterplot(x='total' ,y='pass' , data = dataset)
sns.lineplot(x='total', y= lr.predict(x),data = dataset , color ='red')
plt.show()
```
