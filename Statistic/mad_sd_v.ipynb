```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_1 = np.array([71,65,74,68,72,70])
data_2 = np.array([90,47,43,96,93,51])
no = np.array([1,2,3,4,5,6])

```

```python
mean1 = np.mean(data_1)
mean2 = np.mean(data_2)
mean1,mean2
```

check the mean absolute deviation between data_1 and data_2

```python
mad_1 = np.sum(np.abs(data_1-mean1))/len(data_1)
mad_2 = np.sum(np.abs(data_2-mean2))/len(data_2)
mad_1,mad_2
```

```python
plt.figure(figsize=(12,4))
plt.scatter(x= data_1, y=no ,color= 'blue',label='D1')
plt.scatter(x= data_2, y=no ,color= 'red',label='D2')
plt.axvline(70,color='orange')
plt.legend()
plt.show()
```

**standard deviation**

If the standard deviation is large, it means the data points are spread out more widely from the mean.   
Small SD → most values are close to the average.   
Large SD → values are scattered far above and below the average.

```python
std_2 =np.std(data_2)
std_1 =np.std(data_1)
std_1,std_2
```
