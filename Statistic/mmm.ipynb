```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
ar= np.array([2,4,3,5,7,9,18,12,15,20,2,5,7,2])
np.sum(ar)/len(ar)
```

```python
np.mean(ar)
```

```python
dataset = pd.read_csv("mmm.csv")
dataset.head(3)
```

```python
dataset['sci'].mean()
```

```python
np.mean(dataset['sci'])
```

```python
total_mean = np.mean(dataset["total"])
```

```python
sns.histplot(x='total', data=dataset, bins=[i for i in range (175 ,360,20)])
plt.axvline(total_mean, color='red', linestyle='--', label=f'Mean = {total_mean:.2f}')
plt.legend()
plt.show()
```

MEDIAN   
find the median in data set 

```python
total_median = np.median(dataset['total'])
print(total_median)
```

```python
sns.histplot(x='total', data=dataset, bins=[i for i in range (175 ,360,25)])
plt.axvline(total_mean, color='red', linestyle='--', label=f'Mean = {total_mean:.2f}')
plt.axvline(total_median, color='orange', linestyle=':', label=f'Mean = {total_mean:.2f}')
plt.legend()
plt.show()
```

```python

dataset['total'].value_counts().head(4)
```

```python
total_mode = dataset['total'].mode()[0]
print(total_mode)
```

```python
sns.histplot(x='total', data=dataset, bins=[i for i in range (175 ,360,25)])
plt.axvline(total_mean, color='red', linestyle='--', label=f'Mean = {total_mean:.2f}')
plt.axvline(total_median, color='orange', linestyle=':', label=f'Median = {total_median:.2f}')
plt.axvline(total_mode, color='yellow', linestyle='-.', label=f'Mode = {total_mode:.1f}')
plt.legend()
plt.show()
```
