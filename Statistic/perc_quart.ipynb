**Percentiles and Quantiles**   
A percentile is a measure in statistics that indicates the value below which a given percentage of observations in a dataset falls.   
Percentiles rank data, they donâ€™t measure spread like standard deviation.

The 50th percentile is the median.

The 25th percentile is called the lower quartile (Q1).

The 75th percentile is the upper quartile (Q3).   

IQR = Q3 -Q1   
minimum = Q1-1.5 IQR   
maximum = Q3-1.5 IQR

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('mmm.csv')
print(dataset.head(3))
```

```python
np.percentile(dataset['total'],0)
```

```python
dataset['total'].min()
```

```python
np.percentile(dataset['total'],50) , dataset['total'].median()
```
