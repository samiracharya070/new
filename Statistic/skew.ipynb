```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Skewness**   
it’s a measure that tells us how much the data distribution is asymmetric (not evenly spread) around its mean   

Symmetrical data → Skewness is 0 (like a perfect bell curve in a normal distribution).

Positive skew → Tail on the right is longer; most data are on the left.

Negative skew → Tail on the left is longer; most data are on the right.

```python
dataset =  pd.read_csv('mmm.csv')
dataset.head(3)
```

```python
dataset['math'].skew(),dataset['sci'].skew(),dataset['total'].skew()
```

```python
sns.histplot(x='math',data= dataset,bins=7)
sns.histplot(x='sci',data= dataset,bins=7)
plt.show()
```

```python
sns.histplot(x='total',data= dataset)
plt.show()
```

```python
data = np.random.randint(0,100,100)
skewness_value = pd.Series(data).skew()
skewness_value
```

```python
plt.hist(x=data,bins=20)
plt.show()
```
